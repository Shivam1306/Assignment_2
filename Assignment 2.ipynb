```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
Quotes = []

for page in range(1,5):
    url = f"https://quotes.toscrape.com/page/{page}/"
    response = requests.get(url)
    response = response.content
    soup = BeautifulSoup(response,"html.parser")
    div_ = soup.find('div')
    quotes = div_.find_all('div',class_ = 'quote')
    small_ = soup.find('small')
    author = small_.find_all('small', class_ = 'author')
    
    for quote in quotes:
        spans = quote.find('span').text
        authors = quote.find('small',class_='author').text
        Quotes.append([spans,authors])
```

```python
df = pd.DataFrame(Quotes,columns = ['Quotes','Author'])
```
